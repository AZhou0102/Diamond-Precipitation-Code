```python
import numpy as np

#shell_volume =  4 * np.pi * (radius ** 2) * step_size
#shell_mass = local_density * shell_volume


G = 6.6743015 / (10 ** 11)

def make_density_function(rad, local_densities):
    #print('rad length', len(rad))
    #print('density length', len(local_densities))
    density_function = {rad[i]: local_densities[i] for i in range(len(rad))}
    #print('length of density fucntion',len(density_function))
    #print(density_function)
    return(density_function)


def find_mass_inside(density_function, current_radius):
    step_size = list(density_function.keys())[1] - list(density_function.keys())[0]
    mass_inside = 0
    for r, local_density in density_function.items():
        if r <= current_radius:
            shell_volume = 4 * np.pi * (r ** 2) * step_size
            shell_mass = shell_volume * local_density
            mass_inside += shell_mass
    return(mass_inside)


def adams_williamson(rad, local_densities):
    density_function = make_density_function(rad, local_densities)
    step_size = list(density_function.keys())[1] - list(density_function.keys())[0]
    gravities_and_pressures = {}
    pressures = []
    total_pressure = 0

    for r, local_density in density_function.items():
        if r <= max(density_function.keys()):
            integrand = (G * find_mass_inside(density_function, r) * local_density) / (r ** 2)
            #print('gravity: ', integrand / local_density)
            
            differential_pressure = integrand * step_size
            total_pressure += differential_pressure
            
            #in the dictionary (aka gravities_and_pressures), each r corresponds to a list of [gravity, pressure]
            gravities_and_pressures[r] = [integrand / local_density, total_pressure]
            #print('TOTAL PRESSURE', np.format_float_scientific(total_pressure, 9), 'Pa', '\n')
    
    for r in gravities_and_pressures.keys():
        gravities_and_pressures[r][1] -= gravities_and_pressures[max(density_function.keys())][1]
        gravities_and_pressures[r][1] = -gravities_and_pressures[r][1]
        
    #for x, y in gravities_and_pressures.items():
        #print("Radius: ", x, '\n', 'Gravity: ', round(y[0], 6), '\n', 'Pressure:', np.format_float_scientific(y[1], 9), 'Pa')
        
    #print(f'Pressure at {max(density_function.keys())} meters from the center is {np.format_float_scientific(gravities_and_pressures[max(density_function.keys())][1], 9)} Pa')
    return(gravities_and_pressures)
    #returns dictionary with each radius corresponding to a gravity and a pressure


#moon_radius_list = list(range(100000, 1800000, 100000))
#moon_local_densities = [7700, 7700, 6500, 3400, 3400, 3400, 3400, 3400, 3400, 3400, 3400, 3400, 3400, 3400, 3400, 3400, 3400]
#adams_williamson(moon_radius_list, moon_local_densities)

r_earth = 6.371 * (10 ** 6)
big_radii_list = [*range(1, int(10 * r_earth), int(r_earth / 10))]


#adams_williamson(big_radii_list, nbig_local_densities)
# data taken from https://www.researchgate.net/figure/Pressure-3a-and-Density-3b-within-the-Moon-as-a-function-of-the-radius_fig3_342388891
```
